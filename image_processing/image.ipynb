```python
import PIL.Image
import facets_improc as ip
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = PIL.Image.open('river.png')
img = np.array(h, dtype=np.int64)

img = img[:, :, :3]
img[:, :, 2] = 0
img[:, :, 1] = 0
img[img[:, :, 0] > 210] = 0

plt.imshow(img)
```

```python
W_0 = np.array([[1,1,1],
                [1,1,1],
                [1,1,1]]
              ) * 10

img_linear = ip.linear_filter(img, W_0, mode = "reflect")

for i in range(1):
    img_linear = ip.linear_filter(img_linear, W_0, mode = "reflect")


plt.imshow(img_linear)
```
