```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
images = np.load('cont_output.npy')
cmap = ListedColormap([ 'blue', 'darkgreen', 'darkseagreen', 'grey', 'red'])
boundaries = [-1.1, -0.5, 0.5, 1.1, 2.1, 3.1]  # Boundaries to separate the color intervals
norm = BoundaryNorm(boundaries, cmap.N)  # Use norm to map boundaries to cmap indices
```

```python
plt.imsave('imaging/plotting/cont_1.png', images[0], cmap=cmap)
plt.imsave('imaging/plotting/cont_2.png', images[int(len(images) / 4)], cmap=cmap)
plt.imsave('imaging/plotting/cont_3.png', images[int(2*len(images) / 4)], cmap=cmap)
plt.imsave('imaging/plotting/cont_4.png', images[int(3*len(images) / 4)], cmap=cmap)
plt.imsave('imaging/plotting/cont_5.png', images[-1], cmap=cmap)
```

```python
def rebin(arr, new_shape):
    shape = (new_shape[0], arr.shape[0] // new_shape[0],
             new_shape[1], arr.shape[1] // new_shape[1])
    return arr.reshape(shape).mean(-1).mean(1)
```

```python
d1 = np.load('output.npy')[0]
plt.imshow(d1)
```

```python
d2 = rebin(d1, (100,100))
d2[d2 < 0] = -1
plt.imshow(d2)
```
